## Binary Search P1

```python
def binary_search(seq, elt, right):
    left = 0
    while left<=right: 
        m = (left+right)//2
        if elt > seq[m]:
            left = m + 1
        elif elt < seq[m]:
            right = m - 1
        else:
            return m
    return -1

seq = [int(i) for i in input().split()]
search_seq = [int(i) for i in input().split()]

n = seq[0]
seq = seq[1:]

soln = list()

for i in search_seq[1:]:
    ans = binary_search(seq, i, n-1)
    soln.append(ans)
print(' '.join([str(i) for i in soln]))
```

## Sorting P3

```python
  

import random
def quicksort3(arr, l, r):
    if l + 1 >= r:
        return
    m = random.randint(l, r-1)
    arr[l], arr[m] = arr[m], arr[l]

    m1, m2 = partition3(arr, l, r)

    quicksort3(arr, l, m1)
    quicksort3(arr, m2+1, r)

def partition3(arr, l, r):
    m2 = l
    for i in range(l+1, r):
        if arr[i] <= arr[l]:
            arr[m2+1], arr[i] = arr[i], arr[m2+1]
            m2 += 1

    arr[l], arr[m2] = arr[m2], arr[l]

    m1 = l
    for i in range(l, m2):
        if arr[i] < arr[m2]:
            arr[i], arr[m1] = arr[m1], arr[i]
            m1 += 1
    return m1, m2

n = int(input())
seq = [int(i) for i in input().split()]
quicksort3(seq, 0, n)
for x in seq:
    print(x, end=' ')
```

## Number of Inversions P4

```python
def merge(left, right):
    i, j, inversion_counter = 0, 0, 0
    final = list()
    while i < len(left) and j< len(right):
        if left[i] <= right[j]:
            final.append(left[i])
            i += 1
        else:
            final.append(right[j])
            inversion_counter += len(left) - i
            j += 1

    final += left[i:]
    final += right[j:]
        
    return final, inversion_counter

def mergesort(arr):
    global tot_count
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2

    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])

    sorted_arr, temp = merge(left, right)
    tot_count += temp

    return sorted_arr

tot_count = 0
n = int(input())
seq = [int(i) for i in input().split()]
mergesort(seq)
print(tot_count)
```
