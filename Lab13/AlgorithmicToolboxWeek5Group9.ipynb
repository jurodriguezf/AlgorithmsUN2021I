## Money Search Again P1

```python
import math

def getChange(n):
    denominations = [1, 3, 4]
    minCoins = [0] + [math.inf]*n
    
    for i in range(1, n+1):
        for j in denominations:
            if i >= j:
                coins = minCoins[i-j]+1
                if coins < minCoins[i]:
                    minCoins[i] = coins
    return minCoins[n]
    
if __name__ == '__main__':
    print(getChange(int(input())))
```

## Primitive Calculator P2

```python
from math import inf

func1 = lambda x : x-1
func2 = lambda x : x//2
func3 = lambda x : x//3

def optimal_sequence(n):
    minSequence = [inf] * (n + 1)

    minSequence[0] = 0
    minSequence[1] = 0

    sequence = []
    sequence.append([0])
    sequence.append([1])

    for i in range(2, n+1):
        tmp2 = inf
        tmp3 = inf
        if i % 3 == 0:
            tmp3 = minSequence[func3(i)] + 1
        if i % 2 == 0:
            tmp2 = minSequence[func2(i)] + 1
        tmp1 = minSequence[i - 1] + 1

        minTemp = min(tmp1, tmp2, tmp3)

        minSequence[i] = minTemp
        if minTemp == tmp1:
            sequence.append(sequence[func1(i)] + [i])
            continue
        if minTemp == tmp2:
            sequence.append(sequence[func2(i)] + [i])
            continue
        if minTemp == tmp3:
            sequence.append(sequence[func3(i)] + [i])

    return sequence[-1]


if __name__ == '__main__':
    n = int(input())
    sequence = list(optimal_sequence(n))
    print(len(sequence) - 1)
```

## Edit Distance P3

```python
def editDistance(s, t):
    T = [[None]*(len(s)+1) for _ in range(len(t)+1)]
    # Primera columna = i
    for i in range(len(s)+1):
    	T[0][i] = i
    # Segunda columna = i
    for j in range(len(t)+1):
    	T[j][0] = j
    # Resto de la matriz
    for i in range(1,len(t)+1):
    	for j in range(1,len(s)+1):
    		diff = 0 if s[j-1]==t[i-1] else 1
    		T[i][j] = min(T[i-1][j]+1,T[i][j-1]+1,T[i-1][j-1]+diff)

    return T[i][j]

if __name__ == "__main__":
    print(editDistance(input(), input()))
```
